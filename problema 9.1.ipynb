```python
from numpy import empty, zeros, max
from pylab import imshow,gray, show, hot,inferno
```

```python
#constantes
M=100 #grid squares on a side
v=0.0  
target= 1e-6 #Target accuracy
a=10
eo=1  #epsilon


#creando vectores para guardar los valores depotencial
phi= zeros([M+1,M+1],float)
phi[0,:]=0
rho=zeros([M+1,M+1],float)
rho[20:41,60:81]=1
rho[60:81,20:41]=-1
phiprime=empty([M+1,M+1],float)
#main loop
delta=1.0
while delta> target:
    #calculate new values of the potencial
    for i in range (M+1):
        for j in range(M+1):
            if i==0 or i==M or j==0 or j==M:
                phiprime[i,j]=phi[i,j]
            else:
                phiprime[i,j]= ((phi[i+1,j] + phi[i-1,j]+phi[i,j+1]+phi[i,j-1])/4)+(a**2/4*eo)*rho[i,j]
                
      #calculate maximun difference from old values
    delta= max(abs(phi-phiprime))
                    
     #swap the two arrays around 
    phi, phiprime=phiprime,phi
                
#make a plot
imshow(phi)
hot()
show()
      
```
